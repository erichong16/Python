# **Dataset Variables**

In this project, we decided to analyze and examine a dataset containing the prices and other attributes of almost 54,000 diamonds. The variables included in the dataset are as follows:

**Price:**
price in US dollars (\$326–\$18,823)

**Carat:**
weight of the diamond (0.2–5.01)

**Cut:**
quality of the cut (Fair, Good, Very Good, Premium, Ideal)

**Color:**
diamond colour, from D (best) to J (worst)

**Clarity:**
a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

**X:**
length in mm (0–10.74)

**Y:**
width in mm (0–58.9)

**Z:**
depth in mm (0–31.8)

**Depth:**
total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43–79)

**Table:**
width of top of diamond relative to widest point (43–95)

# **Data Cleaning** 

We imported the file as a CSV from GitHub. From here, we used the python functions of pandas, NumPy, Matplotlib, and Seaborn. The strategy we had used was to analyze the data set first before we did any cleaning. We renamed the length, width, and the depth columns to make more sense when we would use the data set for visualization. To make sure there were no duplicates, we checked by using df.duplicated, thus finding 152 duplicates in total. Therefore, we used the python function df.drop_duplicates to remove them from the dataset. Afterwards, we went on to check for any na values, and there were 32. We had then used the df.dropna to makes sure the na values would be dropped in the correct places.

For visualization purposes, we calculated the width by the diamond's depth using those columns to find the volume. For this dataset, we wanted it to be sorted by price. Thus, we sorted the values ascending from most expensive to least expensive.

```
#import necessary python module(s) and read the csv dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/diamonds.csv',index_col='cut',na_values=0) #make cut column as the index
df = df.loc[:, ~df.columns.str.contains('^Unnamed')] #drop unnamed column
df.describe()
```

```
#see the first 3 rows and columns
df.head(3)
```

```
#see a summary information of the data
df.info()
```

```
#rename the length, width, and depth
df.rename(columns={'depth':'depth%','x':'length','y':'width','z':'depth'},inplace=True)
```

```
#check if there are any duplicates
df.duplicated().sum() #152 duplicates
df.drop_duplicates(inplace=True)
```

```
#drop na values
df.isnull().sum() #32 null values (7 length, 6 width, 19 depth)
df.dropna(subset=['length','width','depth'],inplace=True)
```

```
#calculation to get diamond size/volume
df['diamond_volume']=df.length*df.width*df.depth 
df
```

```
#sort values by price from the most expensive to the least
df.sort_values(by='price',ascending=False,inplace=True)
```

# **Visualization & Aggregation** 

The purpose of this report is to analyze the business-related trends and qualities of diamonds. Many factors correlate with the price of diamonds, but the qualitative questions focused on this report are:
1.  What factors matter the most for the price of a diamond?
2.  What dimensions of the diamond produces a premium to ideal cut?
3.  What are the highest prices for a diamond based on cut and clarity?
4.  What is the mean price range of certain cuts?

The benefit of this analysis will demonstrate to the consumers which qualities of a diamond have the greatest relationship to their overall price of diamonds. The importance of understanding these characteristics will allow the consumer to be a more knowledgeable buyer when in pursuit of what type of diamonds they are interested in.


For the first visualization of the heatmap, we use the cut vs. color as the variables. The colors of colors demonstrate the colors associated here; the lighter the color, the better the quality. In the color analysis, we see how D is the best and J is the worst. However, based on the heatmap results, there is no correlation between the two variables of color vs. cut compared to price. 

```
dfheat=df.pivot_table(index='cut',columns='color',values='price')
fig, ax = plt.subplots(figsize=(12,7))
title = "Cut vs Color Heatmap"
df.index = pd.CategoricalIndex(df.index, categories= ['Fair','Good','Very Good','Premium','Ideal'])
plt.title(title, fontsize=18)
sns.heatmap(dfheat,linewidths=.3,fmt='d')
plt.show()
```

The heatmap visualizes the two variables cut vs. clarity in comparison to the price. When observing the heatmap, the darker the color, the greater the price, but the lighter the color, the smaller the price. However, the clarity and color here show a very low correlation with the price. Leading to the assumption that the types of colors and clarities of a diamond are not closely held to the price. 

```
dfheatclarity=df.pivot_table(index='cut',columns='clarity',values='price')
fig, ax = plt.subplots(figsize=(13,8))
title1 = "Cut vs Clarity Heatmap"
plt.title(title1, fontsize=18)
df.index = pd.CategoricalIndex(df.index, categories= ['Fair','Good','Very Good','Premium','Ideal'])
df.clarity = pd.CategoricalIndex(df.clarity, categories= ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF'])
sns.heatmap(dfheatclarity,linewidths=.3,fmt='d',cmap='YlGnBu')
plt.show()
```

```
#encode categorical variables (color) into numerical values
df['color']=df['color'].astype('category')
df['color_cat']=df['color'].cat.codes
```

```
#encode categorical variables (calrity) into numerical values
df['clarity']=df['clarity'].astype('category')
df['clarity_cat']=df['clarity'].cat.codes
```

```
df_sample=df.sample(n=2000) #get a sample for comparing correlation between price against depth and price against carat
```

This scatterplot demonstrates the price of a diamond vs. the carat of a diamond. Based on the graph below, we can conclude that the carat does have a great coloration with the price. We added the diamond color variable to paint a better picture of what type of diamonds are correlated to the specific carat. The darker the color on the scatterplot is the worst colors and the brightest are the best. Therefore, with every step up in the diamond's carat size, we see a price increase.

```
#see the correlation between price and carat
df['carat'].corr(df.price) #0.9215433258283714
#visualize using scatter plot
fig, ax =plt.subplots()
fig.set_size_inches(10.5, 8.5)
scatter=plt.scatter('carat','price',data=df_sample,c=df_sample.color_cat, s=10)
plt.title('Relationship between Price and Carat')
plt.xlabel('Diamond Carat')
plt.ylabel('Price of Diamond')
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="color")
ax.add_artist(legend1)
```

Throughout this scatterplot, we see that the variables being accounted for are the price vs. the diamond's depth. As in the previous scatterplot, we use the color to demonstrate the diamonds that have the same depth compared to their color. However, the price of the diamond does have a high correlation with the depth of the diamond. Thus, with each step up in the value of the depth, the price increases for the diamond. 

```
#see the correlation between price and depth
df['depth'].corr(df.price) #0.8680221524784041
#visualize using scatter plot
fig, ax =plt.subplots()
fig.set_size_inches(10.5, 8.5)
scatter=plt.scatter('depth','price',data=df_sample,c=df_sample.color_cat, s=10)
plt.title('Relationship between Price and Depth')
plt.xlabel('Diamond Depth')
plt.ylabel('Price of Diamond')
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper left", title="color")
ax.add_artist(legend1)
```

The scatterplot here is using the variables of price vs. the variables x, y, z. These variables represent eth length, width, and hight of the diamond. Thus, we calculated these three variables to get the sum of the size. In scatterplot, we use the color to demonstrate the diamonds that do have the same size in comparison with their color. Hence, teh size of the diamond deos have a correlation with the price. Each step up in sizee of the diamond increases the price.

```
df['diamond_volume'].corr(df.price) #0.904090474440292
fig, ax =plt.subplots()
fig.set_size_inches(10.5, 8.5)
scatter=plt.scatter('diamond_volume','price',data = df_sample,c=df_sample.color_cat, s=10)
plt.title('Relationship between Price and Diamond Size')
plt.xlabel('Diamond Size')
plt.ylabel('Price of Diamond')
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper left", title="color")
ax.add_artist(legend1)
plt.show()
```

The bar graph here is to demonstrate the price's mean vs. the median of the color. This demonstrates with each color (D is the best and J is the worst), there is a price decrease when there is a higher value in color. Therefore, this bar graph demonstrates how there is no correlation between price and color. 

```
#visualize the mean and median of each color against the price
df.pivot_table(values='price',index='color',aggfunc=('mean','median')).plot(stacked=False,kind='bar')
plt.title('Price Mean and Median Against Diamond Colors')
plt.xticks(rotation=0)
plt.ylabel('price')
```

The bar graph here is to demonstrate the price's mean vs. the median of the cut. This demonstrates that with each specific cut of the diamond from best to worst (Fair, Good, Very Good, Premium, Ideal), the mean and median of the price have no fluctuation. Therefore, this bar graph demonstrates how there is no correlation between price and cut.

```
#visualize the mean and median of each cut against the price
df.pivot_table(values='price',index='cut',aggfunc=('mean','median')).plot(kind='bar',color=('green','yellow'))
df.index = pd.CategoricalIndex(df.index, categories= ['Fair','Good','Very Good','Premium','Ideal'])
plt.title('Price Mean and Median Against Diamond Cut')
plt.ylabel('price')
plt.ylim(None,5000)
plt.xticks(rotation=0)
plt.show()
```

This bar graph visualizes the average and median price against the diamond clarity. The I1 is the best clarity, while the IF is the worst. This graph again demonstrates that there is a pretty weak correlation on both variables as the better the clarity, the lowe the price is, in general.

```
#visualize the mean and median of each color against the price
df.pivot_table(values='price',index='clarity',aggfunc=('mean','median')).plot(stacked=False,kind='bar',color=('red','purple'))
df.clarity = pd.CategoricalIndex(df.clarity, categories= ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF'])
plt.title('Price Mean and Median Against Diamond Clarity')
plt.xticks(rotation=0)
plt.ylabel('price')
```
